```python
import pandas as pd
pd.options.display.float_format = '{:.2%}'.format
```

```python
def dividir(x):
    aux= x.split(" ")
    if(len(aux)==1):
        fase= "Octavos"
    else:
        fase= aux[1]
    return fase
```

```python
data= pd.read_csv("DatosSimulacionNew.csv")
```

```python
data["Fase Actual"] = data["Fase alcanzada"].apply(lambda x: dividir(x))
```

```python
data
```

```python
df= pd.DataFrame()
```

## Octavos

```python
df=data[data["Fase Actual"]=="Octavos"][["Ganador", "Simulacion"]].groupby("Ganador").count()
```

```python
df.rename(columns={"Simulacion": "Gano Octavos"},index={"Ganador": "Pais"} ,inplace= True)
```

```python
aux= data[data["Fase Actual"]=="Octavos"][["Perdedor", "Simulacion"]].groupby("Perdedor").count()
```

```python
aux.rename(columns={"Simulacion": "Perdio Octavos"}, index={"Perdedor": "Pais"} ,inplace= True)
```

```python
df["Perdio Octavos"] = aux["Perdio Octavos"]
```

```python
df["Octavos"]=df.apply(lambda x: x["Gano Octavos"]+x["Perdio Octavos"], axis=1)
```
