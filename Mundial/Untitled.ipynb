```python
import pandas as pd
import numpy as np
```

```python
class Selection:
    def __init__(self, name, group,win,draw,lose,status,ranking, odd_group, odd_first):
        self.name= name
        self.group= group
        self.win= win
        self.lose= lose
        self.draw=draw
        self.status= status
        self.ranking= ranking
        self.odd_group= odd_group
        self.odd_first = odd_first
    
    def __str__(self):
        string= self.name +', status: ' + self.status 
        return string
```

```python
data= pd.read_csv("mundial.csv")
```

```python
data.sort_values("Ranking", ascending= True)
```

```python
data["Win"]=data.apply(lambda x: x["LW"]+x["VW"]+(x["WinPenalitation"]*4), axis=1)
data["Draw"]= data.apply(lambda x: x["LD"]+x["VD"]+(x["DrawPenalization"]*4), axis=1)
data["Lose"]= data.apply(lambda x: x["LL"]+x["VL"]+(x["LosePenalitation"]*4), axis=1)
```

```python
def create_selections(data):
    ls_selections= []
    for index, row in data.iterrows():
        name= row["Selecction"]
        group= row["Group"]
        win=row["Win"]
        draw=row["Draw"]
        lose= row["Lose"]
        status= row["Status"]
        ranking= row["Ranking"]
        odd_group=row["GruopOdds"]
        odd_first= row["FirstOdds"]
        selection= Selection(name,group,win,draw,lose,status,ranking,odd_group, odd_first)
        ls_selections.append(selection)
    return ls_selections
```

```python
def get_proba_groups(ls_selections):
    dc_groups= {}
    for i,group in enumerate(list(data["Group"].unique())):
        dc_groups[group]= ls_selections[4*i: 4*(i+1)]

    dc_group_prob={}
    for item in dc_groups.items():
        #print(item[0])
        ls_prob=[]
        ls_results=[]
        for first in item[1]:
            for second in item[1]:
                if(first==second):
                    pass
                else:
                    result= first.name+'-'+ second.name
                    ls_results.append(result)
                    result_prob= 100/(first.odd_first*second.odd_group)
                    #print(f"{result} con probabilidad {result_prob}")
                    ls_prob.append(result_prob/100)
        ls_prob= list(np.array(ls_prob)/sum(ls_prob))
        dc_group_prob[item[0]] = (ls_results,ls_prob)
    return dc_groups, dc_group_prob
```

```python
def get_classified(dc_groups,dc_group_prob):
    dc_classified={}
    for item in dc_group_prob.items():
        prediction_group=np.random.choice(item[1][0], p=item[1][1])
        prediction_group= prediction_group.split("-")
        dc_classified[item[0]+'1']= (prediction_group[0],item[0])
        first=search_selection(dc_groups,item[0],prediction_group[0])
        dc_classified[item[0]+'2']= (prediction_group[1],item[0]) 
        second=search_selection(dc_groups,item[0],prediction_group[1])
        first.status= item[0]+'1'
        second.status=item[0]+'2'
    return dc_classified
```

```python
def search_selection(dc_groups, group, name):
    for selection in dc_groups[group]:
        if(selection.name== name):
            return selection
```

```python
def eliminatorias(knockout_brackets,dc_classified,dc_groups,stage,dataSimulation,simulationNumber):    
    dc_best={}
    for item in knockout_brackets.items():
        teams= item[1].split("-")
        teamA= search_selection(dc_groups,dc_classified[teams[0]][1], dc_classified[teams[0]][0])
        teamB= search_selection(dc_groups,dc_classified[teams[1]][1], dc_classified[teams[1]][0])
        ls_options= [teamA.name,teamB.name,"Draw"]
        ls_prob= get_match_prob(teamA,teamB)
        #print(f"Se juega {teamA} vs {teamB}")
        while(True):
            #print(f"Las probabilidades son: {ls_prob}")
            result= np.random.choice(ls_options,p=ls_prob)
            if(result=="Draw"):
                #print("Hubo empate, reajustando")
                if(teamA.ranking> teamB.ranking):
                    if((ls_prob[1]>0.05) and (ls_prob[2]>0.05)):
                        ls_prob[0]+=.1
                        ls_prob[1]-=.05
                        ls_prob[2]-=.05
                else:
                    if((ls_prob[0]>0.05) and (ls_prob[2]>0.05)):
                        ls_prob[1]+=.1
                        ls_prob[0]-=.05
                        ls_prob[2]-=.05
            else:
                break
        winner=ls_options.index(result)
        if(winner==0):
            teamA.status= item[0] + " "+stage
            dc_best[item[0]]=(teamA.name, teamA.group)
            result= teamA
            lose= teamB
        else:
            teamB.status= item[0] + " "+ stage
            dc_best[item[0]]=(teamB.name, teamB.group)
            result= teamB
            lose=teamA
        X= pd.Series([simulationNumber,result.name, lose.name, lose.status, stage])
        dataSimulation= dataSimulation.append(X,ignore_index=True)
        #print(f"Gano {result}")

    return dc_best,dataSimulation
```

```python
def simulation(data,simulation_number,dataSimulation):
    brackets8={"W49": "A1-B2", "W50": "C1-D2", "W53": "E1-F2", "W54": "G1-H2",
                   "W51": "B1-A2", "W52": "D1-C2", "W55": "F1-E2", "W56": "H1-G2"}
    brackets4={"W57": "W49-W50", "W58":"W53-W54", "W59":"W51-W52", "W60": "W55-W56"}
    brackets2={"W61": "W57-W58", "W62": "W59-W60"}
    winner= {"Winner": "W61-W62"}
    ## Selecciones
    ls_selections= create_selections(data)
    ## Fase de grupos
    dc_groups, dc_group_prob= get_proba_groups(ls_selections)
    dc_classified= get_classified( dc_groups,dc_group_prob)
    ## Octavos
    dc_best8,dataSimulation= eliminatorias(brackets8, dc_classified,dc_groups, "Cuartos",dataSimulation,simulation_number)
    ## Cuartos
    dc_best4,dataSimulation=eliminatorias(brackets4,dc_best8,dc_groups, "Semis",dataSimulation,simulation_number)
    ## Semis
    finals,dataSimulation= eliminatorias(brackets2, dc_best4,dc_groups, "Finales",dataSimulation,simulation_number)
    ## Finales
    champion,dataSimulation= eliminatorias(winner, finals,dc_groups, "Campeon",dataSimulation,simulation_number)
    return champion,dataSimulation
```

```python
def get_match_prob(teamA,teamB):
    odd_wA= teamA.win + teamB.lose
    odd_wB= teamA.lose + teamB.win
    odd_draw= teamA.draw + teamB.draw
    total=odd_draw+odd_wA+odd_wB
    odd_draw/=total
    odd_wA/=total
    odd_wB/=total
    return [odd_wA,odd_wB,odd_draw]
```

```python
dataSimulation=pd.DataFrame(columns=[0,1,2,3,4])
for i in range(3000): 
    champion,dataSimulation= simulation(data,i,dataSimulation)
```

```python
dataSimulation.rename(columns={0:"Simulacion",1: "Ganador", 2:"Perdedor", 3:"Fase alcanzada", 4:"Siguiente fase"}, inplace=True)
```
