# Modulos

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.svm import SVR
```

# Datos

```python
datos= pd.read_csv("/content/Salarios.csv")
X= datos.iloc[:,1:2].values
y= datos.iloc[:,-1].values
```

# Support Vector Machine

## Sin escalar

```python
regresion= SVR(kernel= "rbf")
regresion.fit(X,y)
```

### Predicciòn

```python
y_pred= regresion.predict([[6.5]])
```

### Visualizaciòn

```python
plt.scatter(X,y,color='red')
plt.plot(X,regresion.predict(X),color='blue')
plt.title("SVM")
plt.xlabel("Posiciòn del empleado")
plt.ylabel("Sueldo")
plt.show()
```

## Escalados

```python
from sklearn.preprocessing import StandardScaler
ssx=StandardScaler()
ssy=StandardScaler()
```

```python
X= ssx.fit_transform(X)
y= ssy.fit_transform(y.reshape(-1,1))
```

```python
regresion= SVR(kernel= "rbf")
regresion.fit(X,y)
```

### Predicciòn

```python
y_pred=regresion.predict(ssx.fit_transform([[6.5]]))
```

```python
y_pred= ssy.inverse_transform(y_pred.reshape(-1,1))
```

```python
y_pred
```

### Visualizaciòn

```python
plt.scatter(X,y,color='red')
plt.plot(X,regresion.predict(X),color='blue')
plt.title("SVM escalado")
plt.xlabel("Posiciòn del empleado")
plt.ylabel("Sueldo")
plt.show()
```

```python
X1= ssx.inverse_transform(X)
plt.scatter(X1,ssy.inverse_transform(y),color='red')
plt.plot(X1,ssy.inverse_transform(regresion.predict(X).reshape(-1,1)),color='blue')
plt.title("SVM Valores reales")
plt.xlabel("Posiciòn del empleado")
plt.ylabel("Sueldo")
plt.show()
```
