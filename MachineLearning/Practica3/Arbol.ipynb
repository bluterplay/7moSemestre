# Modulos

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

# Datos

```python
datos= pd.read_csv("/content/Salarios.csv")
X= datos.iloc[:,1:2].values
y= datos.iloc[:,-1].values
```

# Àrbol de regresiòn

```python
regresion= DecisionTreeRegressor(random_state=0)
regresion.fit(X,y)
```

```python
y_pred=regresion.predict([[6.5]])
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,regresion.predict(X),color='blue')
plt.title("Arbol de regresion")
plt.xlabel("Posiciòn del empleado")
plt.ylabel("Sueldo")
plt.show()
```
