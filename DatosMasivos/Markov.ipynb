## Markov

```python
P = [[0,1/3,1/3,1/3], [0,0,.5,.5],[1,0,0,0],[.5,0,.5,0]]
P
```

```python
l,e= np.linalg.eig(np.transpose(P))
l,e
```

```python
eingReal =e[:,0]
eingReal= eingReal/np.sum(eingReal)
```

```python
eingReal
```

### Vector de estados en $t$

```python
r= [.25,.25,.25,.25]
T = np.transpose(P)
rt=np.dot(T,r)
eps= .001
while (np.sum(abs(r-rt))>eps):
    r= rt
    rt= np.dot(T,r)
```
