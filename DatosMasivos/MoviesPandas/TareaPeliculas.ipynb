## Modulos

```python
import pandas as pd
from datetime import datetime
```

## Datos

```python
columns_name=['user_id','item_id','rating','timestamp']
datos=pd.read_csv("u.data",sep="\t",names=columns_name)
```

```python
columns_name = ['item_id', 'title', 'release_date', 'video_release_date', 'imdb_url','unknown', 'Action', 'Adventure',
'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy','Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

movies = pd.read_csv('u.item', sep='|', names=columns_name,encoding='latin-1')
```

## Top 3 mejores y peores

```python
ratings =pd.merge(datos,movies, on= 'item_id')
```

### Mejores

```python
ratings.groupby(['item_id','title']).mean().sort_values('rating', ascending= False).head(3)
```

### Peores

```python
ratings.groupby(['item_id','title']).mean().sort_values('rating', ascending= False).tail(3)
```

## Pelicula con más calificaciones

```python
ratings.groupby(['item_id','title']).count().sort_values('rating', ascending= False).head(1)["user_id"]
```

## Usuario que más ha calificado

```python
ratings.groupby(['user_id']).count().sort_values('rating', ascending= False).head(1)[["rating"]]
```

## Promedio de evaluaciones por pelicula

```python
ratings.groupby(['item_id','title']).count().sort_values('rating', ascending= False)[['rating']] /len(ratings['item_id'].value_counts())
```

## Numero de peliculas por año

```python
ratings["year"] = ratings[["timestamp"]].applymap(lambda x: datetime.fromtimestamp(x).year)
```

```python
ratings.groupby(['year','title']).count().groupby("year").count()["user_id"]
```
